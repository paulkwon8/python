```python
f = open('seoul.csv') #p29
data = csv.reader(f)  #p83
next(data)

result_01 = []  #비어있는 리스트를 준비
result_08 = []  #비어있는 리스트를 준비

# 누락된 기온 ''
for row in data:  #한줄 한줄 꺼낸다.
    if row[-1] !='': #누락되어 있지 않으면
        date = row[0]
        if date.split('-')[1] =='01':
            result_01.append(float(row[-1]))  #최고기온을 실수로 변환해서 저장
        elif date.split('-')[1] == '08':
            result_08.append(float(row[-1]))  #최고기온을 실수로 변환해서 저장
```

```python
result_01, result_08
```

```python
plt.figure(figsize=(3,1.2))
n = plt.hist(result_01, bins=100, color='b')
n = plt.hist(result_08, bins=100, color='r')
```

```python
f = open('seoul.csv')
data = csv.reader(f)
next(data)
result = []

for row in data:
    if row[-1] != '':
        date = row[0].split('-')
        if date[1] == '06' and date[2] =='19':
            result.append(float(row[-1]))        
```

```python
plt.figure(figsize=(3,1.2))
n = plt.hist(result, bins=100, color='r')
```

```python
import pandas as pd
data = pd.read_csv('seoul.csv', encoding='cp949')
data = data.dropna() #전처리 : 누락된 값이 있으면 그 행은 버린다.
data.head()  #상위 5건
```

```python
month =[]
for i in range(12):
    month.append([])
x = np.array(data)
x.shape, x[0]
```

```python
np.mean(x[:,-1])
```

```python
for i in range(len(x)):
    row = x[i]
    date = row[0]
    high = row[-1]
    mon = date.split('-')[1]
    #1월은 0번째 위치에 저장해야 하므로 -1
    month[int(mon)-1].append(high)
```

```python
plt.figure(figsize=(5,1.1))
m = plt.boxplot(month)
```
